```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('tmdb_movies.csv')
genres = pd.read_csv('tmdb_genres.csv')
```

```python
movies
```

```python
quantile3 = movies.quantile(q=0.75)
quantile3['vote_count']
```

```python
movies_quantile3 = movies[movies['vote_count'] > quantile3['vote_count']]
```

```python
movies_by_vote_average = movies_quantile3.sort_values('vote_average', ascending=False)
best10movies = movies_by_vote_average[:10]
best10movies
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies2010_2016 = movies[(movies['release_date'] >= '2010-01-01') & (movies['release_date'] <= '2016-12-31')]
revenue_budget_mean = movies2010_2016.groupby(pd.Grouper(key = 'release_date', freq = 'Y')).agg({'revenue' : 'mean', 'budget' : 'mean'})
revenue_budget_mean.reset_index(inplace = True)
revenue_budget_mean
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(revenue_budget_mean['release_date'], revenue_budget_mean['revenue'], label = 'revenue')
axes.plot(revenue_budget_mean['release_date'], revenue_budget_mean['budget'], label = 'budget')
axes.legend(loc= (1.05, 0.9))
axes.set_title('Mean revenue and budget of movies produced in 2010 - 2016')
```
