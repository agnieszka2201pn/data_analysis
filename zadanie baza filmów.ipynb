```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('tmdb_movies.csv')
genres = pd.read_csv('tmdb_genres.csv')
```

```python
movies
```

```python
quantile3 = movies.quantile(q=0.75)
quantile3['vote_count']
```

```python
movies_quantile3 = movies[movies['vote_count'] > quantile3['vote_count']]
movies_quantile3['vote_count'].min()
```

```python
movies_by_vote_average = movies_quantile3.sort_values('vote_average', ascending=False)
best10movies = movies_by_vote_average[:10]
best10movies
```
