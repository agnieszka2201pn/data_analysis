Script for calculation of savings for purchase of an appartment.

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
#Initial data

appartment_price = 120000
annual_inflation_rate = 0.05
interest_on_the_deposit = 0.12/12 #capitalized monthly
time_until_purchase = 5
months_until_purchase = time_until_purchase * 12
```

```python
#compound interest calcualtion for the final price of the appartment after 5 years

appartment_final_price = np.around(appartment_price*(1+annual_inflation_rate)**5,2)
appartment_final_price
```

```python
monthly_deposit = np.around(- npf.pmt(interest_on_the_deposit, months_until_purchase , pv = 0, fv=appartment_final_price),2)
monthly_deposit
```

```python
# monthly changes of appartment price and deposit value - linear.

appartment_price_changes = np.linspace(appartment_price, appartment_final_price, months_until_purchase)
deposit_value_changes = np.linspace(0, appartment_final_price, months_until_purchase)

plt.plot(appartment_price_changes, label='appartment price')
plt.plot(deposit_value_changes, label ='deposit value')
plt.xlabel('number of months')
plt.ylabel('value')
```

```python
result = monthly_deposit
deposit_changes = []
n=1
while n <= months_until_purchase:
    deposit_changes.append(result)
    result = result + result*interest_on_the_deposit + monthly_deposit
    n+=1

deposit_changes_arr = np.around(np.array(deposit_changes), 2)
deposit_changes_arr

#jak zrobić od razu ndarray, żeby wyszło tak jak w powyższej pętli while?

plt.plot(appartment_price_changes, label='appartment price')
plt.plot(deposit_changes, label ='deposit value')
plt.xlabel('number of months')
plt.ylabel('value')
```
