```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(11)
y = x ** 2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color = 'red')
plt.plot(x, y-10, color = 'blue', ls = 'dashed')
plt.xlabel('oś x')
plt.ylabel('oś Y')
plt.title('mój pierwszy wykres')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='red')
plt.subplot(1,2,2)
plt.plot(y,x, color = 'blue')
```

```python
rand_arr = np.random.randint(1,1000, 2000).reshape(1000,2)
plt.scatter(rand_arr[:, 0], rand_arr[:, 1])
```

```python
cmap=np.empty(rand_arr.shape, dtype='object')
cmap[:] = 'blue'
cmap[rand_arr.min(axis=1)>300] = 'red'
plt.scatter(rand_arr[:, 0], rand_arr[:,1], c=cmap[:, 0])
```

```python
plt.figure(figsize=(10,5))
plt.scatter(rand_arr[:, 0], rand_arr[:,1], c=cmap[:, 0])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, label = 'x^2', color = 'red', linewidth = 3, 
         marker = 'o',
         markersize = 10,
         markerfacecolor = 'orange',
         markeredgewidth = 3)
axes.plot(x, x**3, label = 'x^3', alpha = 0.3, linestyle ='dotted',
         linewidth = 2, marker = '*', markerfacecolor = 'black')
axes.plot(x, x**4, label = 'x^4', marker = 'x', markersize = 2, markeredgewidth = 10)
axes.legend(loc=(1.05, 0.0))
axes.set_xlabel('oś x')
axes.set_ylabel('oś y')
axes.set_title('funny chart')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x, x**2, label = 'x^2', color = 'red', linewidth = 3, 
         marker = 'o',
         markersize = 10,
         markerfacecolor = 'orange',
         markeredgewidth = 3)
axes.plot(x, x**3, label = 'x^3', alpha = 0.3, linestyle ='dotted',
         linewidth = 2, marker = '*', markerfacecolor = 'black')
axes.plot(x, x**4, label = 'x^4', marker = 'x', markersize = 2, markeredgewidth = 10)
axes.legend(loc=(1.05, 0.0))
axes.set_xlabel('oś x')
axes.set_ylabel('oś y')
axes.set_title('funny chart')
axes.yaxis.set_major_locator(plt.NullLocator())
axes.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.bar(np.array(['a'*n for n in range(1,11)]), np.arange(1,11))
axes.xaxis.set_major_locator(plt.MaxNLocator(3))
axes.yaxis.set_major_locator(plt.MultipleLocator(2))
axes.set_ylim()
```
