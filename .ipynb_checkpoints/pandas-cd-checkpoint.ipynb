```python
import pandas as pd
df = pd.read_excel('Pivot.xlsx')
df.head()
```

```python
import numpy as np
df.pivot_table(values='Sprzedaż', index = 'Przedstawiciel', columns = 'Region', aggfunc=np.sum).fillna(0).round(2)
```

```python
df.pivot_table(values='Sprzedaż', index = ['Region', 'Przedstawiciel'], aggfunc = np.sum).round(2)
```

```python
df.pivot_table(values='Sprzedaż', index = 'Region', aggfunc = [len, np.max, np.min])
```

```python
def commission_fee(x):
    if x <= 300:
        return 0
    elif x <= 900:
        return x * 0.03
    else: 
        return x * 0.06
```

```python
df['commission_fee'] = df['Sprzedaż'].apply(lambda x: commission_fee(x))
```

```python
import math
df['#_opakowań'] = df['Sztuki'].apply(lambda x: math.ceil(x/5))
df
```

```python
def bonus(row):
    margin = (row['Sprzedaż'] - row['Koszty'])/row['Sprzedaż']
    if margin > 0.55:
        return 200
    else:
        return 0
```
