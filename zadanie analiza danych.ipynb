Step2: check what RecruitmentSource is the best if the employment length is our priority.

```python
def work_period_calculation(row):
    return (row['DateofTermination'] - row['DateofHire'])
```

```python
HR_data['Work_Period'] = HR_data.apply(lambda row: work_period_calculation(row), axis = 1)
HR_data['Work_Period']
```

```python
HR_data['Work_Period'] = pd.to_numeric(HR_data['Work_Period'].dt.days, downcast='integer')
```

```python
recruitment_source_effectiveness = HR_data.groupby(pd.Grouper(key = 'RecruitmentSource')).agg({'Work_Period': 'mean'}).sort_values(by = 'Work_Period', ascending = False)
recruitment_source_effectiveness.reset_index(inplace = True)
```

```python
recruitment_source_effectiveness.plot.bar(x= 'RecruitmentSource', y='Work_Period', figsize = (15,5))
```

Conclusion: it can be observed that recruitment method may have some influence on the average work period. Carreerbuilder is a most effectives recruitment source in this respect. On the other end we can observe that Company Intranet-Partner, On-line Web application and PayPerClick result in the shortest employment time.

Step 4: age structure of employees

```python
HR_data['DOB']
```

```python
HR_data['yearOfBirth'] = HR_data['DOB'].dt.year
```

```python
HR_data
```

```python
current_year = 2022
def calculate_age(row):
    return (current_year - row['yearOfBirth'])

HR_data['EmployeeAge'] = HR_data.apply(lambda row: calculate_age(row), axis=1)
```

```python
HR_data = HR_data[(HR_data['yearOfBirth']<2001)]
HR_data
```

```python
sns.set_style('darkgrid')
```

```python
sns.distplot(HR_data['EmployeeAge'])
```

```python
HR_data['EmployeeAge'].describe()
```

```python
g = sns.FacetGrid(data = HR_data, col = 'Department', row = 'Sex')
```

```python
g.map(plt.scatter, 'EmployeeAge', 'RaceDesc')
```

Conclusion: Employees' age varies from 30 to 53 years, the median value is 40 years old. It can be concluded that in Admin Offices both male and female employees are below mean age. Female employees in Sales are below mean age, while male employees age varies largely. In both IT and Production age distribution is very wide. Interestingly, in Software Engineering male employees' age is centered around the median.
