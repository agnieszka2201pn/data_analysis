```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
HR_data = pd.read_csv('HRDataset.csv')
HR_data
```

Step 0: Preparation of database, remove empty rows.

```python
plt.figure(figsize = (15, 10))
sns.heatmap(HR_data.isnull())
```

```python
HR_data = HR_data.dropna(thresh = 2)
HR_data
```

Step 1: check if there is any correlation between MangerID and employee PerformanceScore.

```python
sns.distplot(HR_data['PerfScoreID'])
```

```python
figure = plt.figure(figsize = (15, 5))
axes = figure.add_axes([0,0,1,1])
sns.barplot(x='ManagerID', y='PerfScoreID', data = HR_data)
```

```python
figure = plt.figure(figsize = (15, 5))
axes = figure.add_axes([0,0,1,1])
sns.boxplot(x='ManagerID', y='PerfScoreID', data = HR_data)
```

```python
sns.lmplot(x='ManagerID', y='PerfScoreID', data = HR_data)
```

Conclusion: managers rate employees performance in 4-step scale (0-3). For all managers the mean value is around 3. It is also the most common performance rating. There is no clear correlation between ManagerID and employee PerformanceScore.
